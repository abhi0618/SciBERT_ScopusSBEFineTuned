```python
import pandas as pd
from tqdm import tqdm

df = pd.read_excel("ScopusJoournalsAll.xlsx")

with open('pre_train_text.txt','w', encoding="utf-8") as f:
    for i in tqdm(range(len(df))):
        text = df["Abstract"][i] + ' [SEP] ' + df["Title"][i]
        if len(text):
          f.write(text + "\n");
        
```
