```python
import pickle
import pandas as pd
from tqdm.notebook import trange
```

```python
data = pd.read_excel("ScopusJoournalsAll.xlsx")

with open('predictions.pkl', 'rb') as f:
    predictions = pickle.load(f)
```

```python
entities = []
for idx in trange(len(data)): 
    pred = predictions[idx]
    res = []
    for entity in pred:
        if "O" not in entity.values(): 
            res.append(entity)
    entities.append(res)
```

```python
assert len(data) == len(entities)
```

```python
data["Entities"] = entities
data.to_csv("Final.csv", index = False)
data.head()
```
